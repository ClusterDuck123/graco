```python
import os
import pandas as pd

DATA_DIRECTORY = "/media/clusterduck123/joe/data"
RAW_DATA_DIRECTORY = f"{DATA_DIRECTORY}/raw-data"
YEAST_DIRECTORY = f"{DATA_DIRECTORY}/processed-data/yeast"
HUMAN_DIRECTORY = f"{DATA_DIRECTORY}/processed-data/human"
```

```python
def get_maxrun(DIRECTORY):
    splitted_file_names = [name.split('_') for name in os.listdir(DIRECTORY)]
    pre_runs = [int(run) for run, ncluster, db_txt in splitted_file_names]
    if pre_runs:
        return max(pre_runs)
    else:
        return 0
```

# Yeast

## Clusters

```python
CLUSTER_DIRECTORY = f"{YEAST_DIRECTORY}/clusterings"

for feature in os.listdir(CLUSTER_DIRECTORY):
    for distance in os.listdir(f"{CLUSTER_DIRECTORY}/{feature}"):
        maxrun = get_maxrun(f"{CLUSTER_DIRECTORY}/{feature}/{distance}/kmedoid")
        print(feature, distance, maxrun)
    print()
```

# p-values

```python
CLUSTER_DIRECTORY = f"{YEAST_DIRECTORY}/p-values"

for feature in os.listdir(CLUSTER_DIRECTORY):
    for distance in os.listdir(  f"{CLUSTER_DIRECTORY}/{feature}"):
        for aspect in os.listdir(f"{CLUSTER_DIRECTORY}/{feature}/{distance}/kmedoid"):
            maxrun = get_maxrun( f"{CLUSTER_DIRECTORY}/{feature}/{distance}/kmedoid/{aspect}")
            print(feature.ljust(6), distance.ljust(16), aspect, maxrun)
        print()
    print()
```

# Enrichment

```python
ENRICHMENT_DIRECTORY = f"{YEAST_DIRECTORY}/enrichments"

for feature in os.listdir(ENRICHMENT_DIRECTORY):
    for distance in sorted(os.listdir(          f"{ENRICHMENT_DIRECTORY}/{feature}")):
        for aspect in os.listdir(        f"{ENRICHMENT_DIRECTORY}/{feature}/{distance}/kmedoid"):
            for correction in os.listdir(f"{ENRICHMENT_DIRECTORY}/{feature}/{distance}/kmedoid/{aspect}"):
                with open(               f"{ENRICHMENT_DIRECTORY}/{feature}/{distance}/kmedoid/{aspect}/"
                                         f"{correction}/clusters.csv", 'r') as f:
                    l1 = len(f.readline().strip(',\n').split(','))
                with open(               f"{ENRICHMENT_DIRECTORY}/{feature}/{distance}/kmedoid/{aspect}/"
                                         f"{correction}/GO-terms.csv", 'r') as f:
                    l2 = len(f.readline().strip(',\n').split(','))
                with open(               f"{ENRICHMENT_DIRECTORY}/{feature}/{distance}/kmedoid/{aspect}/"
                                         f"{correction}/genes.csv", 'r') as f:
                    l3 = len(f.readline().strip(',\n').split(','))

                print(feature.ljust(6), distance.ljust(16), aspect, correction, (l1+l2+l3)/3)
        print()
    print()
```
