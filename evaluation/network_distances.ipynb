```python
from pathlib import Path
from itertools import combinations
from scipy.spatial.distance import squareform
from sklearn.metrics import precision_recall_curve, auc, roc_curve, roc_auc_score

import numpy as np
import pandas as pd
import seaborn as sns
import ringity as rng
import networkx as nx
import matplotlib.pyplot as plt
```

```python
NETWORK_DIR = "/home/clusterduck123/Desktop/transfere/networks/"
MODEL_DIR = NETWORK_DIR + '/model_networks'
#REAL_DIR  = NETWORK_DIR + '/real-world_networks'

models = [directory.stem for directory in Path(MODEL_DIR).glob('*') if directory.is_dir()] 
#models = [directory.stem for directory in Path(REAL_DIR).glob('*') if directory.is_dir()] 
```

```python
GDV_df = pd.read_csv(f"/home/clusterduck123/Desktop/transfere/network_distances/synt_GCD11.csv", 
            header = [0,1],
            index_col=[0,1])

GDV_y_scores = 1-squareform(GDV_df, checks=False)
GDV_y_true   = pd.DataFrame(np.zeros(GDV_df.shape, dtype=int), 
                        columns = GDV_df.columns, 
                        index   = GDV_df.index)

for mod in models:
    GDV_y_true.loc[mod,mod] = 1

GDV_y_true = squareform(GDV_y_true, checks=False)

GDV_prec, GDV_reca, GDV_thre = precision_recall_curve(y_true=GDV_y_true, probas_pred=GDV_y_scores)
GDV_AUC = auc(GDV_reca, GDV_prec)

GDV_fpr, GDV_tpr, GDV_thre = roc_curve(y_true=GDV_y_true, y_score=GDV_y_scores)
GDV_ROC_AUC = roc_auc_score(y_true=GDV_y_true, y_score=GDV_y_scores)
```

```python
DAGO_df = pd.read_csv(f"/home/clusterduck123/Desktop/transfere/network_distances/synt_avg_DAGO_jsdi.csv", 
            header = [0,1],
            index_col=[0,1])

DAGO_y_scores = 1-squareform(DAGO_df, checks=False)
DAGO_y_true   = pd.DataFrame(np.zeros(DAGO_df.shape, dtype=int), 
                        columns = DAGO_df.columns, 
                        index   = DAGO_df.index)

for mod in models:
    DAGO_y_true.loc[mod,mod] = 1

DAGO_y_true = squareform(DAGO_y_true, checks=False)
DAGO_prec, DAGO_reca, DAGO_thre = precision_recall_curve(y_true=DAGO_y_true, probas_pred=DAGO_y_scores)
DAGO_AUC = auc(DAGO_reca, DAGO_prec)

DAGO_fpr, DAGO_tpr, DAGO_thre = roc_curve(y_true=DAGO_y_true, y_score=DAGO_y_scores)
DAGO_ROC_AUC = roc_auc_score(y_true=DAGO_y_true, y_score=DAGO_y_scores)
```
